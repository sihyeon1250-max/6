## 1-1. 이 이진 검색 트리의 리프 노드는 총 몇 개인가?

**리프 노드의 수:** 자식 노드가 없는 노드로 **7, 17, 70** 총 **3개**입니다.

## 1-2. 현재 높이 4를 유지하면서 키를 추가한다면 가능한 리프 노드의 최대 개수는 몇 개인가?

**높이 4일 때 최대 리프 노드 수:**

높이가 $h$일 때 포화 이진 트리의 리프 노드 수는 $2^{h-1}$입니다.

높이가 4이므로 $2^{4-1} = \mathbf{8}$**개**입니다.

## 1-3. 높이 4인 이진 검색 트리가 포화 이진 트리 모양이 되면 리프 노드와 내부 노드의 수는 몇 개나 차이가 나는가?

**포화 이진 트리에서 리프 노드와 내부 노드의 수 차이:**

| 구분 | 수식 | 값 (h=4) |
|------|------|-----------|
| 리프 노드 수 | $2^{h-1}$ | 8개 |
| 내부 노드 수 | $2^{h-1} - 1$ | 7개 |

따라서 차이는 항상 **1**입니다.

## 1-4. 위 3)의 성질은 높이가 $h$인 모든 포화 이진 트리에 대해 성립하는가?

**일반화:** 네, 높이가 $h$인 **모든** 포화 이진 트리에서 리프 노드의 수는 내부 노드의 수보다 항상 **1이 더 많습니다.**

$$\text{리프 노드 수} - \text{내부 노드 수} = 2^{h-1} - (2^{h-1} - 1) = 1$$

## 2-1. 트리의 총 노드 수가 $n$이면 리프 노드의 수는 몇 개인가?

**완전 이진 트리에서 리프 노드 수:**

총 노드 수가 $n$일 때, 리프 노드의 수는 $\lceil n/2 \rceil$ ($n/2$를 올림한 값)입니다.

## 2-2. 위 1)에서 리프 노드의 수와 내부 노드의 수는 어떤 관계인가?

**관계:** 리프 노드의 수는 내부 노드의 수와 **같거나 1만큼 더 많습니다.**

> 이진 트리의 기본 성질: $L = I + 1$ (포화/완전 이진 트리에서 성립)

## 3-1. 전위 순회 (Preorder Traversal)

**순회 방식:** Root → Left → Right

$$5 \to 3 \to 3 \to 7 \to 6 \to 8 \to 9$$

## 3-2. 중위 순회 (Inorder Traversal)

**순회 방식:** Left → Root → Right

$$3 \to 3 \to 6 \to 7 \to 5 \to 8 \to 9$$

## 3-3. 후위 순회 (Postorder Traversal)

**순회 방식:** Left → Right → Root

$$3 \to 6 \to 7 \to 3 \to 9 \to 8 \to 5$$

## 3-4. 중위 순회가 항상 정렬된 순서를 보장함을 증명하시오.

**증명 (귀납법):**

이진 검색 트리(BST)의 성질:
- 임의의 노드 $v$에 대해, **왼쪽 서브트리의 모든 키 < $v$.key < 오른쪽 서브트리의 모든 키**

중위 순회는 `왼쪽 서브트리 → 루트 → 오른쪽 서브트리` 순으로 방문하므로:

1. 왼쪽 서브트리를 중위 순회하면 루트보다 작은 값들이 정렬되어 출력됨
2. 루트를 방문
3. 오른쪽 서브트리를 중위 순회하면 루트보다 큰 값들이 정렬되어 출력됨

이를 재귀적으로 적용하면, **중위 순회는 항상 오름차순 정렬된 순서**로 키를 방문합니다. $\blacksquare$

## 5. 키 16 삽입 후의 트리 모양

**삽입 경로:**

55(왼쪽) → 15(오른쪽) → 28(왼쪽) → 18(왼쪽) → **16 삽입**

> **결과:** 18의 **왼쪽 자식 노드**로 16이 추가됩니다.

## 6. 키 38 삭제 후의 트리 모양

**삭제 케이스:** 자식이 하나인 노드

- 38은 자식이 하나(33)인 노드입니다.
- 38을 삭제하고, **33을 48의 왼쪽 자식**으로 연결합니다.

## 7. 키 3 삭제 후의 트리 모양

**삭제 케이스:** 리프 노드

- 3은 **리프 노드**입니다.
- 단순히 연결을 끊어 삭제합니다.

## 8. 키 15 삭제 후의 트리 모양

**삭제 케이스:** 자식이 둘인 노드

두 가지 방법 중 하나를 선택:
- 왼쪽 서브트리의 **최댓값(8)** 을 올리거나
- 오른쪽 서브트리의 **최솟값(18)** 을 올림 ← 일반적으로 사용

> **결과:** 18을 15 위치로 올리고, 기존 18 자리의 노드들을 재배치합니다.

## 9. $n$개의 원소로 구성되는 이진 검색 트리의 최대 높이와 최소 높이

| 구분 | 트리 형태 | 높이 |
|------|----------|------|
| **최대 높이** | 편향 트리 (Skewed Tree) | $n$ |
| **최소 높이** | 완전 이진 트리 | $\lfloor \log_2 n \rfloor + 1$ |

- **최대 높이 $n$:** 노드가 한쪽 방향으로만 연결된 편향 트리일 때
- **최소 높이 $\lfloor \log_2 n \rfloor + 1$:** 완전 이진 트리 형태로 균형 잡혔을 때
