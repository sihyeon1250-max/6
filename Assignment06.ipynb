1-1. 이 이진 검색 트리의 리프 노드는 총 몇개인가?

리프 노드의 수: 자식 노드가 없는 노드로 7, 17, 70 총 3개입니다.

1-2. 현재 높이 4를 유지하면서 키를 추가한다면 가능한 리프 노드의 최대 개수는 몇 
개인가?

높이 4일 때 최대 리프 노드 수: 높이가 h일 때 포화 이진 트리의 리프 노드 수는 2^{h-1}입니다. 높이가 4이므로 2^{4-1} = 8개입니다.

1-3. 높이 4인 이진 검색 트리가 포화 이진 트리 모양이 되면 리프 노드와 리프 노드를 
제외한 나머지 노드(내부 노드라 한다)의 수는 몇 개나 차이가 나는가?

포화 이진 트리에서 리프 노드와 내부 노드의 수 차이:
리프 노드 수: 2^{h-1}
내부 노드 수: 2^{h-1} - 1
따라서 차이는 항상 1입니다. (8개 vs 7개)

1-4. 위 3)의 성질은 높이가 ℎ인 모든 포화 이진 트리에 대해 성립하는가?

일반화: 네, 높이가 h인 모든 포화 이진 트리에서 리프 노드의 수는 내부 노드의 수보다 항상 1이 더 많습니다.

2-1. 트리의 총 노드 수가 𝑛이면 리프 노드의 수는 몇 개인가?

완전 이진 트리에서 리프 노드 수: 총 노드 수가 n일 때, 리프 노드의 수는 [n/2](n/2를 올림한 값)입니다.

2-2. 위 1)에서 리프 노드의 수와 내부 노드의 수는 어떤 관계인가?

관계: 리프 노드의 수는 내부 노드의 수와 같거나 1만큼 더 많습니다. (정확히는 L = I + 1 관계는 포화/완전 트리에서 성립하는 이진 트리의 기본 성질입니다.)

3-1.이 검색 트리를 전위 순회 방식으로 순회할 때 노드의 방문 순서를 나열하시오. 
순서는 노드에 있는 킷값으로 적으시오. 

전위 순회 (Root-L-R): 5 → 3 → 3 → 7 → 6 → 8 → 9

3-2.이 검색 트리를 중위 순회 방식으로 순회할 때 노드의 방문 순서를 나열하시오.

중위 순회 (L-Root-R): 3 → 3 → 6 → 7 → 5 → 8 → 9

3-3.이 검색 트리를 후위 순회 방식으로 순회할 때 노드의 방문 순서를 나열하시오.

후위 순회 (L-R-Root): 3 → 6 → 7 → 3 → 9 → 8 → 5

3-4.중위 순회는 모든 이진 검색 트리에 대해 항상 키를 정렬된 순서로 방문한다는 
사실을 증명하시오.

증명 요약: 중위 순회는 왼쪽 서브트리(작은 값) → 루트 → 오른쪽 서브트리(큰 값) 순으로 방문하므로, 재귀적 정의에 의해 항상 오름차순 정렬된 순서가 됩니다.

4.다음 이진 검색 트리에서 키 30을 검색하는과정을 순서대로 설명하시오.

루트 55와 비교: 30 < 55이므로 왼쪽 자식(15)으로 이동.
15와 비교: 30 > 15이므로 오른쪽 자식(28)으로 이동.
28와 비교: 30 > 28이므로 오른쪽 자식(30)으로 이동.
30 발견: 검색 성공.

5.04번 문제에 이진 검색 트리에서 키 16이 삽입된 후의 모양을 그리시오.

루트 55보다 작음(왼쪽) → 15보다 큼(오른쪽) → 28보다 작음(왼쪽) → 18보다 작음(왼쪽).

결과: 18의 왼쪽 자식 노드로 16이 추가됩니다.

6.다음 이진 검색 트리에서 키 38을 삭제한 후의 모양을 그리시오.

38은 자식이 하나(33)인 노드입니다. 38을 지우고 그 자리에 33을 끌어올려 48의 왼쪽 자식으로 연결합니다.

7. 06번 문제에서 주어진 이진 검색 트리에서 키 3을 삭제한 후의 모양을 그리시오.

3은 리프 노드입니다. 단순히 연결을 끊어 삭제합니다.

8.06번 문제에서 주어진 이진 검색 트리에서 키 15를 삭제한 후의 모양을 그리시오.

자식이 둘인 노드입니다. 왼쪽 서브트리의 가장 큰 값(8) 또는 오른쪽 서브트리의 가장 작은 값(18)을 올립니다. 보통 오른쪽의 최소값인 18을 15 위치로 올리고, 기존 18 자리에 있던 노드들을 재배치합니다.

9.𝑛개의 원소로 구성되는 이진 검색 트리의 가능한 최대 높이와 최소 높이는 어떻게 
되는가?

최대 높이: 노드가 한쪽으로만 치우친 편향 트리(Skewed Tree)일 때이며, 높이는 n입니다.
최소 높이: 완전 이진 트리 형태일 때이며, 높이는 [log_2 n] + 1 입니다.
